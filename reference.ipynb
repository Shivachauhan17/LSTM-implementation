```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from  gensim.models import Word2Vec
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=["the glass of milk",
      "the glass of juice",
      "the cup of tea",
      "I am a good boy",
      "I am a good developer",
      "understand the meaning of words",
      "your videos are good",
      ]

```

# one hot repr in vocubaliries

```python
voc_size=10000
```

```python
onehot_repr=[one_hot(words,voc_size) for words in sent]
len(onehot_repr)
```

# make equal size of all one hot encodings

```python
from keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences #all sentence should have the equal words->good Embadding matrix
from keras.models import Sequential
```

```python
sent_length=8
embedded_doc=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_doc)
```

```python
#number of feature for each word
dim =10  
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_doc))
```
